```python
import pandas as pd
finance_1 = pd.read_csv("/content/Finance_1.csv")
finance_1
```

```python
finance_2 = pd.read_csv("/content/Finance_2.csv")
finance_2
```

```python
#KPI 1 = YEAR WISE LOAN AMOUNT
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
finance_1['issue_year'] = pd.to_datetime(finance_1['issue_d']).dt.year
sns.barplot(x= 'issue_year', y = 'loan_amnt', data = finance_1, palette='coolwarm')
plt.title('YEAR WISE LOAN AMOUNT')
plt.xlabel('year')
plt.ylabel('loan amount')
plt.show()
```

```python
#KPI 2 - GRADE AND SUB-GRADE WISE REVOL BALANCE
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
merged_df = pd.merge(finance_1, finance_2, on='id', how='inner')
merged_df
revol_bal_summary = merged_df.groupby(['grade', 'sub_grade'])['revol_bal'].sum().reset_index()
revol_bal_summary
revol_bal_summary = revol_bal_summary.sort_values(by=['grade', 'sub_grade'])
plt.figure(figsize=(18,6))
plt.bar(revol_bal_summary['sub_grade'], revol_bal_summary['revol_bal'])
plt.title('GRADE AND SUB-GRADE WISE REVOL BALANCE')
plt.xlabel('Sub-grade')
plt.ylabel('Total Revolving Balance')
plt.show()
```

```python
#KPI 3 - TOTAL PAYMENT FOR VERIFIED AND NON-VERIFIED STATUs
df = merged_df[['verification_status', 'total_pymnt']]
pymnt = df.groupby('verification_status')['total_pymnt'].sum()
plt.figure(figsize=(6,6))
plt.pie(pymnt, labels=pymnt.index, autopct='%1.1f%%')
plt.title('VERIFICATION STATUS WISE TOTAL PAYMENT')
plt.show()
```

```python
#KPI 4 - STATE WISE MONTH WISE LOAN STATUS
merged_df['issue_date'] = pd.to_datetime(merged_df['issue_d'], errors='coerce')

# Extract month name for grouping
merged_df['issue_month'] = merged_df['issue_date'].dt.strftime('%b')

# Grouping: State + Month + Loan Status
state_month_status = merged_df.groupby(['addr_state', 'issue_month', 'loan_status']).size().reset_index(name='count')

print("STATE-WISE + MONTH-WISE + LOAN STATUS COUNT:")
display(state_month_status)

pivot_table = state_month_status.pivot_table(
    values='count',
    index='addr_state',
    columns='loan_status',
    aggfunc='sum',
    fill_value=0
)

pivot_table.plot(kind='bar', figsize=(18,6))
plt.title("State Wise Loan Status Distribution")
plt.xlabel("State")
plt.ylabel("Number of Loans")
plt.legend(title="Loan Status")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# KPI 5 - HOME OWERSHIP WISE LAST PAYMENT DATE
# Convert last payment date column to datetime
merged_df['last_pymnt_d'] = pd.to_datetime(merged_df['last_pymnt_d'], errors='coerce')

# Grouping: Home Ownership vs Latest Last Payment Date
home_last_payment = merged_df.groupby('home_ownership')['last_pymnt_d'].max().reset_index()

print("Home Ownership Wise - Last Payment Date:")
display(home_last_payment)

plt.figure(figsize=(8,4))
plt.bar(home_last_payment['home_ownership'], home_last_payment['last_pymnt_d'].astype(str))
plt.title('Home Ownership vs Last Payment Date')
plt.xlabel('Home Ownership')
plt.ylabel('Last Payment Date')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
